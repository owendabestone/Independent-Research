The first clustering was done with k=3 (three clusters) since in the Perera paper, they found k=3 to be most effective. After running a PCA and plotting the clusters on the first two principal components, we noticed that cluster 2 and 3 each occupied a quadrant of the graph while cluster 1 covered the samples in the second and third quadrant. Because of the imbalance in the areas covered by clusters, I changed the number of clusters to four.

### Interpretion/ conclusion 


The findings tell us three things: One, students have to spend a certain amount of time to receive good grades. The less confident and less comfortable a student is with coding, the more time he/she should spend on projects and homework. Two, Cognitive Science majors are generally less prepared than Data Science and Computer Science majors. This finding parallels with the major curriculums as this is a coding/data science focused class and CS/DS majors are more likely to have previously been exposed to those subjects more than COGS students have. This sigals that COGS majors might need more assitance than CS/DS majors. Three, the average letter grade for every cluster is an A. This suggests that COGS108 grading is very lenient. If this were a harder class, we might see bigger grade differences between the clusters. 

There is one particular outlier that caught our eyes. Student No.214 spent over 130 hours on homework, yet still received a poor grade. This student presented very low confidence in programming as well as statistics. He/she is a typical student who needs the most help with. Since we only have one such case, we cannot make a promising recommendation to this type of student. But one thing we do notice is that Student No.214 had no indication of using Piazza (the class forum). It is possible that the student was not aware of the existence of a class forum that could have helped with understanding the concepts to some extent. 

To promote an effective student learning experience, we recommend students to spend the average amount of time cluster 1 and 2 spent on this class. If the student has a strong coding and data science background, we advise the student to spend roughly 40 hours studying/working outside of the class. As for students with a weak background (often are the Cognitive Science majors), we advise them to have 60 hours of outside-of-classroom studying.  

### Future Study

However, this study provides a good framework on how to find hidden patterns in a class data using a clustering analysis.
